```python
import gensim.downloader as api

# Carregar embeddings GloVe pré-treinados
glove_model = api.load("glove-wiki-gigaword-100")

# Obter vetor de palavra para uma palavra específica
word_vector = glove_model['cellphone']
print("Vetor GloVe para a palavra 'cell phone':", word_vector)

# Encontrar palavras mais similares
similar_words = glove_model.most_similar('cellphone', topn=5)
print("Palavras mais similares a 'cellphone':", similar_words)
```
